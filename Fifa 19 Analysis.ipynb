# Initial data cleaning 

We can see that there are many null values, we can remove the columns that have too many null values.
We can deal with the other null values in different ways, either impute them using median/mean/mode or remove them 

```python
#We will remove the loaned from column as ~93% of the column is missing values 
```

```python
df1 = df.copy()
```

```python
df1 = df1.drop('Loaned From', axis =1)
```

I will also drop all the other rows with missing values as the maximum values we can lose is 11% which will still result in a dataset of ~14700 records 

```python
df1.shape
```

```python
df1 = df1.dropna()
```

```python
df1.shape
```

I will also drop the unamed column 

```python
df1 = df1.drop('Unnamed: 0', axis =1)
```

Lets identify unnecessary columns that can be dropped 

```python
cols_to_drop = ['ID', 'Photo', 'Flag', 'Club Logo']
```

```python
df1 = df1.drop(cols_to_drop, axis=1)
```

```python
df1.head(1)
```

Lets remove the '+2' values on the positions for the players 

```python
positions = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
```

```python
for col in positions:
    df1[col] = df1[col].apply(lambda x: str(x).split('+')[0])
```

```python
df1.head(1)
```

As we are predicting the overall score of a player, I believe it will be predicated on their stats and not their wage, value, release clause, Real Face, Jersey Number, contract validity or the special column 

```python
df1 = df1.drop(['Value', 'Wage', 'Special', 'Real Face', 'Jersey Number','Contract Valid Until', 'Release Clause'], axis=1)
```

```python
df1.head(1)
```

```python
df1.Joined.unique()
```

Lets work out how long the player has been playing for - this data is from 2019 so we will subtract the start of the players career year from 2019 

```python
#Extract the year from the joined column
```

```python
df1.Joined = df1.Joined.apply(lambda x: int(str(x).split(',')[1]))
```

```python
df1.Joined.unique()
```

```python
df1['Years of experience'] = 2019 - df1.Joined
```

```python
df1.drop('Joined', axis=1, inplace=True)
```

```python
df1.head(1)
```

Lets review the work rate

```python
df1['Work Rate'].value_counts()
```

This can be encoded using One Hot Encoding later

As our target varaible as 'Overall' I do not believe the other positions the player can play will be extremely useful, as we want to build our predictive model on player playing in their OWN positions. For example we will only want to build a model using Lionel Messi's data when he is playing in RF which is his intented position to maximize his overall stat.
Therefore, we can drop them. 

```python
#Drop unneeded position columns
df1.drop(['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM',
       'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB',
       'LCB', 'CB', 'RCB', 'RB'], axis=1, inplace=True)
```

```python
df1.head(1)
```

The final part of cleaning the data can come from converting height into cm so that it is in a continuous scale, and removing the lb from weights 

```python
df1.Weight = df1.Weight.apply(lambda x: int(str(x).strip('lbs')))
```

- Feet to cm conversion = 30.48
- Inches to cm conversion = 2.54

```python
df1['Feet'] = df1.Height.apply(lambda x: int(str(x).split("'")[0]))
df1['Inches'] = df1.Height.apply(lambda x: int(str(x).split("'")[1]))
```

```python
df1['Feet'] = df1['Feet']*30.48
df1['Inches'] = df1['Inches']*2.54
```

```python
df1['Height in cm'] = df1['Feet'] + df1['Inches']
```

```python
df1.drop(['Feet', 'Inches'], axis=1, inplace=True)
```

```python
df1.drop('Height', axis=1, inplace=True)
```

```python
df1 = df1.rename(columns={'Weight':'Weight in lbs'})
```

```python
df1.head(1)
```

```python
df1['Body Type'].value_counts()
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) >= threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_cols = correlation(df1, 0.85)
corr_cols
```

# Data preprocessing 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, SCORERS
from sklearn.ensemble import RandomForestRegressor
```

Lets first split our data into our X and y datasets 

```python
X = df_final.drop('Overall', axis=1)
y = df_final['Overall']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```
